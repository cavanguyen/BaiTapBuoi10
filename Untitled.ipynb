```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
df_train
```

```python
df_train['px']=df_train['px_height']*df_train['px_width']
df_train['sc']=df_train['sc_h']*df_train['sc_h']
df_test['px']=df_train['px_height']*df_train['px_width']
df_test['sc']=df_train['sc_h']*df_train['sc_h']
```

```python
df_train
```

```python
feature_columns=['battery_power','clock_speed','fc','int_memory','m_dep','mobile_wt','n_cores','pc','px','ram','sc','talk_time','blue','dual_sim','four_g','three_g','touch_screen','wifi']
label_columns="price_range"

X_train=df_train.loc[:,feature_columns]
y_train=df_train.loc[:,label_columns]
X_test=df_test.loc[:,feature_columns]
y_test=df_test.loc[:,label_columns]
```

from sklearn.preprocessing import StandardScaler, MinMaxScaler
min_max_scaler=MinMaxScaler()
X_train[['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep',
       'mobile_wt', 'n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_h',
       'sc_w', 'talk_time']] = min_max_scaler.fit_transform(X_train[['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep',
       'mobile_wt', 'n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_h',
       'sc_w', 'talk_time']])

```python
from sklearn.svm import SVC

classifier = SVC(C=1.0, break_ties=False, cache_size=200, class_weight=None, coef0=0.0,
    decision_function_shape='ovr', degree=3, gamma='scale', kernel='linear',
    max_iter=-1, probability=False, random_state=None, shrinking=True,
    tol=0.001, verbose=True)
classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)

from sklearn.metrics import accuracy_score
```

```python
print("Accuracy_score:", accuracy_score(y_test, predictions))
```
